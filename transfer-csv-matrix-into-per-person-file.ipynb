```python
import csv
from pandas import DataFrame
from pprint import pprint

with open("books-responses.csv", encoding='utf-8') as fp:
    reader = csv.DictReader(fp)
    rows = [line for line in reader]
    
    for row in rows:
        if '\ufeffYear' in row:
            row["Year"] = row.pop("\ufeffYear")
        elif '\ufeffBook' in row:
            row["Book"] = row.pop("\ufeffBook")
        if "" in row:
            row.pop("")
    
df = DataFrame(rows)

df
```

Want to get all book selections for a given person and save as text file

```python
from pprint import pprint

person = "Nastya Ambr"

def get_book_pics_file(person):
    return "data/raw-picks/{}.txt".format(person.replace(" ", "_"))

def save_book_picks_for_person(person):
    books = []
    for _, row in df[df[person] == "x"].iterrows():
        books.append(row["Book"])
    print("Person: %s" % person)
    print("Books:")
    pprint(books)
    fname = get_book_pics_file(person)
    with open(fname, "w") as fp:
        for book in books:
            fp.write(book + "\n")
            
save_book_picks_for_person(person)
```

Try to do the above for every single person

```python
import os

for header in df:
    if header not in ["Book", "Year", "Author", "Votes in Common"]:
        fname = get_book_pics_file(header)
        if os.path.exists(fname):
            print("WARNING: over-writing file for %s" % header)
        
        save_book_picks_for_person(header)
```
